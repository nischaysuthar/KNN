```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[df.target==0]
df.head()
```

```python
df[df.target==1].head()
```

```python
df1=df[:50]
df2=df[50:100]
df3=df[100:]
```

```python
import matplotlib.pyplot as plt

plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',label='setosa')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',label='setosa')
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',label='setosa')

plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',label='setosa')
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(['target','flower_name'],axis='columns')
y=df.target

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.predict([[2.5,3.5,4.5,5.5]])
```

```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
df=pd.DataFrame(digits.data,digits.target)
df
```

```python
df['target']=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(['target'],axis='columns')
y=df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
